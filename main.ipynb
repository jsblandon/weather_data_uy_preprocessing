```python
import pandas as pd
from pandas import DataFrame
from IPython.display import display
from weather_data_preprocessing import column_preci_treatment, data_preci_treatment, null_report
```

```python
# Data directory
dir = '/directory/'

# Filenames
columns_route = dir + 'PrecipitacionMensualLocalidadesColumnasAdicionales(1981-2010).csv'
data_route = dir + 'PrecipitacionMensualLocalidadesDatos(1981-2010).csv'

# Data import and preprocessing
df_preci_columns = column_preci_treatment(columns_route)
df_preci_data = data_preci_treatment(data_route)

# Data merge from columns and precipitation data
df_preci_data = df_preci_data.merge(df_preci_columns, how='inner', on='Nombre Estación')

# Null count per variable
display(null_report(df_preci_data))

# Null value dropping
df_preci_data.dropna(inplace=True)
print("Number of entries after dropping null values: " + str(len(df_preci_data)))

# Data storage (uncomment to use it)
# df_preci_data.to_csv(dir + 'PrecipitacionMensualLocalidadesPreprocesada(1981-2010).csv', index=False)
```
